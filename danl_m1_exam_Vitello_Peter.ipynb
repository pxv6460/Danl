**Question 1**

Find the top 5 beer markets in terms of the number of households that purchased beer.
Find the top 5 beer markets in terms of the amount of total beer consumption.

```python
import pandas as pd
path = 'https://bcdanl.github.io/data/beer_markets.csv'
beer_mkts = pd.read_csv(path)

beer_mkts
# find top 5 beer markets based on the number of households that purchased beer
top_househld = beer_mkts.groupby('market')['hh'].sum().nlargest(5)

print(top_househld)

# top 5 beer markets in terms of total consumption
top_comsum = beer_mkts.groupby('market')['beer_floz'].sum().nlargest(5)

print('-------------------------------------------')

print(top_comsum)
```

**Question 2**

For households that purchased BUD LIGHT at least once, what fraction of households did purchase only BUD LIGHT?

For households that purchased BUSCH LIGHT at least once, what fraction of households did purchase only BUSCH LIGHT?

For households that purchased COORS LIGHT at least once, what fraction of households did purchase only COORS LIGHT?

For households that purchased MILLER LITE at least once, what fraction of households did purchase only MILLER LITE?

For households that purchased NATURAL LIGHT at least once, what fraction of households did purchase only NATURAL LIGHT?

Which beer brand does have the largest proportion of such loyal consumers?

```python
# filter the dataframe to find households that purchased the brand of beer at least once
bud_house = beer_mkts[beer_mkts['brand'] == 'BUD LIGHT']['hh'].unique()
busch_house = beer_mkts[beer_mkts['brand'] == 'BUSCH LIGHT']['hh'].unique()
coors_house = beer_mkts[beer_mkts['brand'] == 'COORS LIGHT']['hh'].unique()
miller_house = beer_mkts[beer_mkts['brand'] == 'MILLER LITE']['hh'].unique()
natural_house = beer_mkts[beer_mkts['brand'] == 'NATURAL LIGHT']['hh'].unique()

# calculate the fraction of hh that only purchased one brand by subtracting the other brands then dividing by the brand we are trying to calculate for
bud_only = len(set(bud_house) - set(busch_house) - set(coors_house) - set(miller_house) - set(natural_house)) / len(set(bud_house)) #if len(set(bud_house)) != 0 else 0
busch_only = len(set(busch_house) - set(bud_house) - set(coors_house) - set(miller_house) - set(natural_house)) / len(set(busch_house)) #if len(set(busch_house)) != 0 else 0
coors_only = len(set(coors_house) - set(bud_house) - set(busch_house) - set(miller_house) - set(natural_house)) / len(set(coors_house)) #if len(set(coors_house)) != 0 else 0
miller_only = len(set(miller_house) - set(bud_house) - set(busch_house) - set(coors_house) - set(natural_house)) / len(set(miller_house)) #if len(set(miller_house)) != 0 else 0
natural_only = len(set(natural_house) - set(bud_house) - set(busch_house) - set(coors_house) - set(miller_house)) / len(set(natural_house)) #if len(set(natural_house)) != 0 else 0


print(bud_only, busch_only, coors_only, miller_only, natural_only)

# find the brand with the largest proportion of loyal consumers
loyal_consumers_fract = {
    'Bud Light': bud_only,
    'Busch Light': busch_only,
    'Coors Light': coors_only,
    'Miller LITE': miller_only,
    'Natural Light': natural_only
}

max_loyal = max(loyal_consumers_fract, key=loyal_consumers_fract.get)
print(max_loyal)
```

by using the max function I can get the largest fraction out of the 5 brands and by using my calcuations I can get the households that only bought one brand

**Question 4**

Calculate the log difference between mean revenues in each group of search_stays_on. (This is the log of the average revenue in group of search_stays_on == 1 minus the log of the average revenue in group of search_stays_on == 0.)

```python
import pandas as pd
import numpy as np
paidsearch = pd.read_csv('https://bcdanl.github.io/data/paidsearch.csv')

# Convert date column to datetime format so it reads properly
paidsearch['date'] = pd.to_datetime(paidsearch['date'])

# groupby date and search on then calculate the mean revenue
mean_rev_date = paidsearch.groupby(['search_stays_on', 'date'])['revenue'].mean().fillna(0).unstack()

#print(mean_rev_date)
#print(mean_rev_date.columns)

# Calculate the log difference
log_diff = np.log(mean_rev_date['2012-04-01']) - np.log(mean_rev_date['2012-04-02'])

print(log_diff)
```

```python
import pandas as pd
import numpy as np

def calculate_log_difference(data, revenue_col='revenue', treatment_col='search_stays_on', date_col='date'):
    """
    This function calculates the log difference between mean revenues for each group for every date.

    Args:
        data: A pandas DataFrame containing columns with revenue and a treatment group identifier.
        revenue_col: The name of the column containing revenue values (defaults to 'revenue').
        treatment_col: The name of the column containing the treatment group identifier (defaults to 'search_stays_on').

    Returns:
        A pandas Series containing the log difference for each date.
    """
    # Group data by date and treatment group
    grouped_data = data.groupby([date_col, treatment_col])

    # Calculate mean revenue for each treatment group for every date
    mean_revenue = grouped_data[revenue_col].mean().unstack().fillna(0)

    print("Mean Revenue:")
    print(mean_revenue)

    # Calculate log difference
    log_difference = np.log(mean_revenue.diff())

    # Replace inf and -inf with NaN
    log_difference = log_difference.replace([np.inf, -np.inf], np.nan)

    return log_difference

paidsearch = pd.read_csv('https://bcdanl.github.io/data/paidsearch.csv')

log_diff = calculate_log_difference(paidsearch, revenue_col='revenue', treatment_col='search_stays_on', date_col='date')

print("Log Difference:")
print(log_diff)
```

by defining my own I am able to make custom columns for date mean revenue and search says on and I can calculate the log difference by using the .diff function

```python
import pandas as pd
NFL2023_stuffs = pd.read_csv('https://bcdanl.github.io/data/NFL2023_stuffs.csv')
NFL2023_epa = pd.read_csv('https://bcdanl.github.io/data/NFL2023_epa.csv')

# Merge the two DataFrames on common columns
NFL2023_stuffs_EPA = pd.merge(NFL2023_stuffs, NFL2023_epa[['play_id', 'passer', 'receiver', 'epa']], on='play_id', how='left')

# Remove NA observations in passer
NFL2023_stuffs_EPA = NFL2023_stuffs_EPA.dropna(subset=['passer'])


NFL2023_stuffs_EPA
```
