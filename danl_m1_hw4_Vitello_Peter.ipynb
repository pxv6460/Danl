## Question 1

How many employees have a “Base Salary” within the top 10% of the DataFrame?

```python
import pandas as pd
nyc_payroll = pd.read_csv('https://bcdanl.github.io/data/nyc_payroll_2022_2023.csv')
ninith_percentile = nyc_payroll['Base Salary'].quantile(0.9)

#filters employees with salary in top 10 %
tenth_percentile = nyc_payroll[nyc_payroll['Base Salary'] >= ninith_percentile]

num_employ = len(tenth_percentile)

print(num_employ)
```

using the quantile method I can calcuate the 90th percentile then by using the >= operator I can get the top 10 pecent of employees

## Question 2

Filter the DataFrame for employees who have “OT Hours” greater than 0 but less than 100, and their “Leave Status as of June 30” is “ACTIVE”.

## Question 3

Find the unique job titles in the “DEPARTMENT OF EDUCATION ADMIN” agency and count how many there are.

```python
import pandas as pd
nyc_payroll = pd.read_csv('https://bcdanl.github.io/data/nyc_payroll_2022_2023.csv')

admin_data = nyc_payroll[nyc_payroll['Agency Name'] == 'DEPARTMENT OF EDUCATION ADMIN']

# get unique values for title discription
job_titles = admin_data['Title Description'].unique()

print(len(job_titles))

for title in job_titles:
  print(title)
```

this code uses the unique function to get the total number of different job titles there are in the data frame

## Queston 4

Identify the employee(s) with the highest “Total OT Paid” in the DataFrame.
Include their “First Name”, “Last Name”, and “Total OT Paid”.

```python
# finds the employee with the largests number of overtime paid
ht_ot_paid_empl = nyc_payroll.nlargest(1, 'Total OT Paid')

print(ht_ot_paid_empl[['First Name','Last Name', 'Total OT Paid']])
```

Using the nlargest function we are able to find the employee with the most total paid overtime

## Question 5

Find the five most popular songs.
A value of a song is defined as a combination of a artist_name value and a track_name value.
Who are artists for those five most popular songs?

```python
spotify = pd.read_csv('https://bcdanl.github.io/data/spotify_all.csv')

spotify['song'] = spotify['artist_name'] + "---" + spotify['track_name']

count = spotify['song'].value_counts()

# gets the top 5 items in count
top5 = count.head(5)

for song, count in top5.items():
  print(song)
```

using value counts I can get the count of all the songs plays
