**Question 1** For each party, calculate the total funding contributions that were disbursed or provided for mitigation projects for each year.

```python
import pandas as pd
path = 'https://bcdanl.github.io/data/climate_finance.csv'
climate_finance = pd.read_csv(path)

# filter the data frame for just mitigation projects
mit_projs = climate_finance[climate_finance['Type of support'] == 'mitigation']

# use groupby to combine party and year variables then use reset index
total_funding = mit_projs.groupby(['Party', 'Year'])['Contribution'].sum().reset_index()

total_funding

#print(total_funding)
```

by filtering the dataframe to use only mitigation projects we can then use the groupby function to group the party and the year and get the total contibutions for the party

**Question 2**

How many parties have provided positive funding contributions to other countries or regions for their adaptation projects for every single year from 2011 to 2018?

```python
#party_count = party_group['Year'].agg('value_counts')

# filter the dataframe using and operator to make sure contributions are more than 0
adaptation_projects = climate_finance[(climate_finance['Type of support'] == 'adaptation') & (climate_finance['Contribution'] > 0)]

# group number of unique years and party
party_counts = adaptation_projects.groupby('Party')['Year'].value_counts().reset_index(name='Counts')

years = len(range(2011,2019))

prty_years = party_counts.groupby('Party')['Counts'].sum() == years

num_prty_years = prty_years.sum()

print(num_prty_years)
```

```python
# Filter for adaptation projects with positive contributions
positive_adaptation_projects = climate_finance[(climate_finance['Type of support'] == 'adaptation') & (climate_finance['Contribution'] > 0)]

# Group by party
grouped_by_party = positive_adaptation_projects.groupby('Party')

# Aggregate to count the number of years for each party with positive contributions
# had to look this part up
party_counts = grouped_by_party['Year'].agg(lambda x: len(set(x)))

# Get the total number of years from 2011 to 2018
total_years = len(range(2011, 2019))

# Filter parties that have contributed for all years from 2011 to 2018
parties= party_counts[party_counts == total_years]

# Get the count of such parties
num = len(parties)

print(num)
```

By grouping by party then aggregate the group by unique years in the data I can get the range of years then fliter that set to find parties that have contributed in a positive way

**Question 3**

What are the minimum, first quartile, median, thrid quartile, maximum, mean, and standard deviation of Close and Volume for each company?

```python
path = 'https://bcdanl.github.io/data/stock_2019_2024.csv'
stock = pd.read_csv(path)

# group the stock dataframe by company and use the describe functin
stats = stock.groupby('company')[['Close', 'Volume']].describe()

print(stats.round(3))
```

the describe function shows a bunch of data in the dataframe if you groupby the right varaibe which in this case is company

**Question 4**

Find the 10 largest values for Volume for each company using a lambda function.

```python
# Define a lambda function to find the 10 largest values for Volume
find_l_volume = lambda x: x.nlargest(10, 'Volume')

l_v_company = stock.groupby('company').apply(find_l_volume)

# reset index to make it a dataframe
l_v_company = l_v_company.reset_index(drop=True)

l_v_company
```

this code gives us the 10 largets volume values for every company in the dataframe then using the reset_index function I am able to turn the data back into a dataframe

**Question 5**

Calculate the Z-score of Close for each company on each Date
.
The formula for the Z-score of Close for each company on Date
 is as follows:
Z= (close - mean(close))/std(close)


where

Close = is a company’s Close price on date t;
mean(close) = is the mean value of Close for a company;
std(close) = is the standard deviation of Close for a company.

```python
# calulate the mean and std for the close var for every company
mn_close_comp = stock.groupby('company')['Close'].transform('mean')
std_close_comp = stock.groupby('company')['Close'].transform('std')

#calculate zscore
stock['close_z_sr'] = (stock['Close'] - mn_close_comp) / std_close_comp

print(stock['close_z_sr'])
```

using groupby I can get the values for the close mean and the close std then I can plug all the information into the formula provided by the question

**Question 6**

Use the transform() method on the stock data to represent all the values of Open, High, Low, Close, Adj Close, and Volume in terms of the first date in the data.

To do so, divide all values for each company by the values of the first date in the data for that company.

```python
# Define a function to divide each column by the values of the first date for each company
def normalize_by_first_date(x):
    return x / x.iloc[0]

norm_stock = stock.groupby('company').transform(normalize_by_first_date)

print(norm_stock)
```

by defining a new fuction I can divide each column by the first date for every company then by normalizing the data I can get the correct values
